```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

cal_housing_sk = fetch_california_housing(as_frame=True)
cal_housing = pd.DataFrame(cal_housing_sk.frame)
cal_housing.head()
```

```python
cal_housing.describe()
```

```python
import matplotlib.pyplot as plt

cal_housing.hist(bins=50, figsize=(12,12))
plt.tight_layout()
plt.savefig('images/feat_histograms.pdf')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12,10))
ax = sns.heatmap(cal_housing.corr(), annot=True, cmap="YlGnBu")
plt.yticks(rotation=0) 
plt.show()
```

```python
plt.scatter(cal_housing["AveOccup"], cal_housing["MedHouseVal"])
plt.show()
```

```python
cal_housing_filtered = cal_housing[cal_housing["AveOccup"] < 100]
plt.scatter(cal_housing_filtered["AveOccup"], cal_housing_filtered["MedHouseVal"])
plt.xlabel("AveOccup")
plt.ylabel("MedHouseVal")
plt.tight_layout()
plt.savefig("images/AveOccup_scatter.pdf")
plt.show()
```
