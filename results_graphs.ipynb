```python
import matplotlib.pyplot as plt
from math import *
from matplotlib.colors import ListedColormap

classes = ["dnn", "gp", "rf50", "knn8", "mlr"]
run_times = [50.9, 284, 3.33, 0.412, 0.006]
log_run_times = [log(i) for i in run_times]
mae = [0.321, 0.321, 0.338, 0.427, 0.532]
values = [0,1,2,3,4]
colors = ListedColormap(['r','b','g','c','m'])

scatter = plt.scatter(log_run_times, mae, c=values, cmap=colors)
plt.xlabel("log(run time)")
plt.ylabel("MAE ($100,000)")
plt.legend(handles=scatter.legend_elements()[0], labels=classes)
plt.tight_layout()
plt.savefig('images/run_time_results.pdf')
plt.show()
```
